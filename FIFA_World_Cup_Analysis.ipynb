```python
import pandas as pd

# Load the datasets
matches_df = pd.read_csv('WorldCupMatches.csv')
players_df = pd.read_csv('WorldCupPlayers.csv')
cups_df = pd.read_csv('WorldCups.csv')

# Display the first few rows of each dataset
print("WorldCupMatches.csv:")
print(matches_df.head())
print("\nWorldCupPlayers.csv:")
print(players_df.head())
print("\nWorldCups.csv:")
print(cups_df.head())
```

```python
#Data Cleaning
```

```python
# Check for missing values
print("Missing values in WorldCupMatches.csv:")
print(matches_df.isnull().sum())

print("Missing values in WorldCupPlayers.csv:")
print(players_df.isnull().sum())

print("Missing values in WorldCups.csv:")
print(cups_df.isnull().sum())

# Check for duplicates
print("Duplicates in WorldCupMatches.csv:")
print(matches_df.duplicated().sum())

print("Duplicates in WorldCupPlayers.csv:")
print(players_df.duplicated().sum())

print("Duplicates in WorldCups.csv:")
print(cups_df.duplicated().sum())

# Drop duplicates if any
matches_df.drop_duplicates(inplace=True)
players_df.drop_duplicates(inplace=True)
cups_df.drop_duplicates(inplace=True)

#Verify the data afer cleaning
print("\nData after cleaning:")
print("WorldCupMatches.csv shape:", matches_df.shape)
print("WorldCupPlayers.csv shape:", players_df.shape)
print("WorldCups.csv shape:", cups_df.shape)
```

```python
#Exploratory Data Analysis
#Identifying Key Insights
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Number of matches per World Cup
plt.figure(figsize=(12, 6))
sns.countplot(data=matches_df, x='Year')
plt.title('Number of Matches per World Cup')
plt.xticks(rotation=45)
plt.savefig('World_Cup_matches.png')
plt.show()
```

```python
#Goals scored in each World CUp
plt.figure(figsize=(12, 6))
sns.barplot(data=cups_df, x='Year', y='GoalsScored')
plt.title('Goals Scored in Each World Cup')
plt.xticks(rotation=45)
plt.savefig('Goals_scored.png')
plt.show()
```

```python
#Average attendance per World Cup
plt.figure(figsize=(12, 6))
sns.barplot(data=cups_df, x='Year', y='Attendance')
plt.title('Average Attendance per World Cup')
plt.xticks(rotation=45)
plt.savefig('Average_attendance.png')
plt.show()
```

```python
#Goals Scored by Country
goals_by_country = matches_df.groupby('Home Team Name')['Home Team Goals'].sum() + matches_df.groupby('Away Team Name')['Away Team Goals'].sum()
goals_by_country = goals_by_country.sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=goals_by_country.values, y=goals_by_country.index)
plt.title('Goals Scored by Country')
plt.savefig('Goals_scored_by_country.png')
plt.show()
```

```python
#Home vs Away team performance
home_goals = matches_df['Home Team Goals'].sum()
away_goals = matches_df['Away Team Goals'].sum()
labels = ['Home Goals', 'Away Goals']
sizes = [home_goals, away_goals]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Home vs Away Team Performance')
plt.axis('equal')
plt.savefig('Home_vs_Away_team_performance.png')
plt.show()
```

```python
#Export Cleaned data to CSV Files
matches_df.to_csv('Cleaned_WorldCupMatches.csv', index=False)
players_df.to_csv('Cleaned_WorldCupPlayers.csv', index=False)
cups_df.to_csv('Cleaned_WorldCups.csv', index=False)
```
